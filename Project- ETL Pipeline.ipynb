## Transform

```python
data['Quantity Ordered'].unique()
```

```python
data = data[~(data['Quantity Ordered'] == 'Quantity Ordered')]
```

```python
data['Quantity Ordered']= data['Quantity Ordered'].astype('int')
```

```python
data['Price Each'].unique()
```

```python
data['Price Each'] = data['Price Each'].astype('float')
```

```python
data['total_sales'] = data['Quantity Ordered'] * data['Price Each']
```

```python
data['total_sales']
```

```python
data['order_year'] = pd.DatetimeIndex(data['Order Date']).year
data['order_month'] = pd.DatetimeIndex(data['Order Date']).month
```

```python
data
```

```python
data['Product'].unique()
```

```python
# Group by month and product, and calculate the sum of sales amount
result = data.groupby([data['order_month'], 'Product'])['total_sales'].sum().reset_index()
```

```python
result
```

```python
#create a separate DataFrame
final_sales_data = result.copy()
```
